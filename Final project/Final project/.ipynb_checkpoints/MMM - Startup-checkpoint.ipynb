# Clean the data

```python
df = pd.read_csv("Data/Advertising_Analyst_Work_1.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
# Clean dates
df['Date']= pd.to_datetime(df['Date'])
```

```python
# Clean 
df['Spend'] = df['Spend'].str.replace('$', '')
df['Sales'] = df['Sales'].str.replace('$', '')
df['Advertised SKU Sales'] = df['Advertised SKU Sales'].str.replace('$', '')
df['Other SKU sales'] = df['Other SKU sales'].str.replace('$', '')
```

```python
df 
```

```python
df['Spend'] = pd.to_numeric(df['Spend'], errors='coerce')
df['Advertised SKU Sales'] = pd.to_numeric(df['Advertised SKU Sales'], errors='coerce')
df['Other SKU sales'] = pd.to_numeric(df['Other SKU sales'], errors='coerce')
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
```
